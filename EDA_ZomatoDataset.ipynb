```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
```

```python
ratings
```

```python
# observations

# 1. when rating is between 4.5 to 4.9 --> Excellent
# 2. when rating is between 4.0 to 4.5 --> Very Good
# 3. when rating is between 3.5 to 4.0 --> Good
# 4. when rating is between 3.0 to 3.5 --> Average
# 5. below 3.0 we have poor ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Counts', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Counts', hue='Rating color', data=ratings, palette=['pink','red','orange','yellow','green','blue'])
```

```python
# observation

# 1. not rated count very high
# 2. maximum rating between 2.5 and 4.0
```

```python
# countplot

sns.countplot(x='Rating color', data=ratings, palette=['pink','red','orange','yellow','green','blue'])
```
